# Logistic Regression Exercise

## Problem: Student Admission Prediction

A university wants to predict whether a student will be admitted based on their test scores.

**Dataset**: 
- Feature 1 (`x1`): Math exam score (0-100)
- Feature 2 (`x2`): Verbal exam score (0-100)
- Target (`y`): Admission decision (1 = admitted, 0 = not admitted)



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Student admission data
# Each row: [Math Score, Verbal Score, Admitted (1) or Not (0)]
data = np.array([
    [85, 90, 1],
    [78, 82, 1],
    [92, 88, 1],
    [88, 85, 1],
    [70, 75, 1],
    [95, 92, 1],
    [82, 78, 1],
    [75, 80, 1],
    [90, 87, 1],
    [68, 72, 0],
    [55, 60, 0],
    [48, 52, 0],
    [62, 58, 0],
    [58, 65, 0],
    [45, 50, 0],
    [52, 55, 0],
    [65, 68, 0],
    [50, 48, 0],
    [42, 45, 0],
    [38, 42, 0]
])

# Separate features and target
X = data[:, :2]  # Math and Verbal scores
y = data[:, 2]   # Admission decision

print("Dataset shape:", X.shape)
print("Number of admitted students:", np.sum(y == 1))
print("Number of rejected students:", np.sum(y == 0))
```

### Visualize the data 

```python
# Plot the data
plt.figure(figsize=(8, 6))
admitted = (y == 1)
rejected = (y == 0)

plt.scatter(X[admitted, 0], X[admitted, 1], c='green', marker='o', label='Admitted', s=100)
plt.scatter(X[rejected, 0], X[rejected, 1], c='red', marker='x', label='Not Admitted', s=100)

plt.xlabel('Math Score')
plt.ylabel('Verbal Score')
plt.title('Student Admission Data')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
